# Decision Tree Regression

## Importing the libraries

```
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset

```
dataset = pd.read_csv('Leftbank_Allyear_Coordinate.csv')
x = dataset.iloc[:, :-2].values
y = dataset.iloc[:, -2].values
```

```
print(x)
```

```
y = y.reshape(len(y),1)
print(y)
```

#Splitting the dataset into the Training set and Test set

```
# Assuming x and y are your feature and target variables
total_samples = len(x)
train_ratio = 0.88734  # Percentage of data for training (1 - test_ratio)
test_ratio = 1 - train_ratio

# Calculate the number of samples for training and testing
num_train_samples = int(total_samples * train_ratio)
num_test_samples = total_samples - num_train_samples

# Split the dataset into training and testing sets
x_train = x[:num_train_samples]
y_train = y[:num_train_samples]
x_test = x[num_train_samples:]
y_test = y[num_train_samples:]
```

```
print(x_train)
```

```
print(x_test)
```

##***Training different Models on training set***

## Training the Decision Tree Regression model on the training dataset

```
# prompt: training the decision tree regression model on training set

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x_train, y_train)

#prompt: predict the y

y_pred = regressor.predict(x_test)
print(y_pred)
```

```
from tabulate import tabulate
from termcolor import colored
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# Calculate evaluation metrics
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test,y_pred)

# Calculate y_test_mean
y_test_mean = np.mean(y_test)

# Calculate the numerator
numerator = np.sum((y_test - y_pred)**2)

# Calculate the denominator
denominator = np.sum((y_test - y_test_mean)**2)

# Calculate NSE
NSE = 1 - (numerator / denominator)

# Calculate PBIAS
PBIAS = np.sum(y_test - y_pred) / np.sum(y_test)

# Create a list of lists to represent the data in a tabular form
data = [
    ["Mean Squared Error (MSE):", f"{mse:.0f}"],
    ["Root Mean Squared Error (RMSE):", f"{rmse:.0f}"],
    ["R-squared (R^2):", f"{r2*100:.0f}%"],
    ["Nash-Sutcliffe Efficiency (NSE):", f"{NSE*100:.0f}%"],
#     ["Percent Bias (PBIAS):", f"{PBIAS * 100:.2f}%"],
    ["Mean Absolute Error (MAE)", f"{mae:.0f}"]
]

# Define table headers
headers = ["Metric", "Value"]

# Format the data into a tabular form
table = tabulate(data, headers, tablefmt="fancy_grid")

# Color the table and print it
colored_table = colored(table, "black")
print(colored_table)
```
